```python
import pandas as pd

df = pd.read_csv("data.csv")

df.info(verbose = True)
```

```python
df.head()
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
#df["price_in_euro"] = df["price_in_euro"].str.replace(",", "").astype(float)
```

```python

# Show rows where price_in_euro contains something that's not a digit, dot or comma
mask = ~df["price_in_euro"].astype(str).str.fullmatch(r"[\d.,]+")
df_invalid = df[mask]

print(df_invalid.shape)
df_invalid.head(50)

```

```python
# Filter out rows where 'price_in_euro' is NOT numeric
df_clean = df[df["price_in_euro"].astype(str).str.fullmatch(r"[\d.,]+")]
df_clean = df[df["price_in_euro"].astype(str).str.fullmatch(r"[\d.,]+")].copy()
df_clean["price_in_euro"] = df_clean["price_in_euro"].str.replace(",", "").astype(float)
df_clean.info(verbose = True)
```

```python
df_clean.head(50)
```

```python
print("Original:", len(df))
print("Cleaned:", len(df_clean))
print("Removed:", len(df) - len(df_clean))
```

```python
# Check how many missing (null) values there are
missing_mileage = df_clean["mileage_in_km"].isna().sum()
print(f"Missing values in mileage_in_km: {missing_mileage}")
```

```python
df_clean = df_clean.dropna(subset=["mileage_in_km"])
missing_mileage = df_clean["mileage_in_km"].isna().sum()
print(f"Missing values in mileage_in_km: {missing_mileage}")
```

```python
missing_fuel = df_clean["fuel_consumption_l_100km"].isna().sum()
print(f"Missing values in fuel_consumption_l_100km: {missing_fuel}")
```

```python
df_clean["fuel_consumption_l_100km"].dropna().unique()[:20]
```

```python
# Step-by-step cleaning
# Step 1: Extract numeric values from string (and ignore junk)
df_clean["fuel_consumption_l_100km"] = (
    df_clean["fuel_consumption_l_100km"]
    .astype(str)
    .str.extract(r"([\d,]+)")     # only extract numeric part like '7,2'
    .replace(",", ".", regex=True)  # replace comma with dot
)

# Step 2: Convert to float (safely)
df_clean["fuel_consumption_l_100km"] = pd.to_numeric(
    df_clean["fuel_consumption_l_100km"], errors="coerce"
)
print(df_clean["fuel_consumption_l_100km"].head(10))
print(df_clean["fuel_consumption_l_100km"].isna().sum())
print(f"Total rows in df_clean: {len(df_clean)}")
```

```python
# Filter rows where fuel_type is 'Electric'
evs = df_clean[df_clean["fuel_type"].str.lower() == "electric"]

# Check how many EVs exist
print(f"Total EVs: {len(evs)}")

# Check how many EVs have missing fuel consumption
missing_evs = evs["fuel_consumption_l_100km"].isna().sum()
print(f"EVs with missing fuel consumption: {missing_evs}")

# Show unique values in EV fuel consumption (excluding nulls)
print("Unique fuel consumption values for EVs:")
print(evs["fuel_consumption_l_100km"].dropna().unique())
```

```python
# Filter rows where fuel_type is 'Electric'
ev_rows = df_clean[df_clean["fuel_type"].str.lower() == "electric"]

# Display first 50 full EV rows
pd.set_option("display.max_columns", None)  # Show all columns
pd.set_option("display.max_rows", 50)       # Show up to 50 rows

ev_rows.head(50)
```

```python
import numpy as np

# STEP 1: Create a boolean mask for EVs
is_ev = df_clean["fuel_type"].str.lower() == "electric"

# STEP 2: Extract numeric km values from 'fuel_consumption_g_km' if it's an EV
# Use regex to capture numbers before 'km Reichweite'
df_clean.loc[is_ev, "ev_range_km"] = (
    df_clean.loc[is_ev, "fuel_consumption_g_km"]
    .str.extract(r"(\d+)\s*km", expand=False)
    .astype(float)
)

# STEP 3: Inspect results
print("EV range (in km) - preview:")
print(df_clean.loc[is_ev, "ev_range_km"].dropna().head(10))

# STEP 4 (Optional): Verify how many EVs now have usable range data
ev_with_range = df_clean.loc[is_ev, "ev_range_km"].notna().sum()
total_evs = is_ev.sum()
print(f"EVs with range data: {ev_with_range} out of {total_evs}")
df_clean["ev_has_range_data"] = df_clean["ev_range_km"].notna()
```

```python
df_clean["has_fuel_consumption"] = df_clean["fuel_consumption_l_100km"].notna()
df_clean.head(10)
```

```python
# Unique values preview (excluding NaN)
print(df_clean["fuel_consumption_g_km"].dropna().unique()[:150])
```

```python
# STEP 1: Work only on non-EVs
is_not_ev = df_clean["fuel_type"].str.lower() != "electric"

# STEP 2: Clean 'fuel_consumption_g_km' for non-EVs
df_clean.loc[is_not_ev, "fuel_consumption_g_km_clean"] = (
    df_clean.loc[is_not_ev, "fuel_consumption_g_km"]
    .str.extract(r"(\d+)", expand=False)  # extract numeric part
    .astype(float)
)

# STEP 3 (Optional): Flag whether valid CO₂ data exists
df_clean["has_emissions_data"] = df_clean["fuel_consumption_g_km_clean"].notna()

# Total non-EV cars
total_non_evs = (df_clean["fuel_type"].str.lower() != "electric").sum()

# Non-EVs with valid emissions data
valid_emissions = df_clean["fuel_consumption_g_km_clean"].notna().sum()

print(f"Total non-EVs: {total_non_evs}")
print(f"Non-EVs with valid CO₂ data: {valid_emissions}")
print(f"Percentage with valid CO₂ data: {valid_emissions / total_non_evs:.2%}")
```

```python
# Force conversion to numeric, set errors='coerce' to turn invalid strings into NaN
df_clean["power_kw"] = pd.to_numeric(df_clean["power_kw"], errors="coerce")
df_clean["power_ps"] = pd.to_numeric(df_clean["power_ps"], errors="coerce")


# Check for missing values
missing_kw = df_clean["power_kw"].isna().sum()
missing_ps = df_clean["power_ps"].isna().sum()

print(f"Missing values in power_kw: {missing_kw}")
print(f"Missing values in power_ps: {missing_ps}")

# Quick range check for outliers
print(f"\npower_kw range: {df_clean['power_kw'].min()} to {df_clean['power_kw'].max()}")
print(f"power_ps range: {df_clean['power_ps'].min()} to {df_clean['power_ps'].max()}")
```

```python
# Drop rows where both power_kw and power_ps are missing
df_clean = df_clean.dropna(subset=["power_kw", "power_ps"])

# Confirm it's done
print(f"New shape of dataset: {df_clean.shape}")
```

```python
# Convert registration_date to datetime
df_clean.loc[:, "registration_date"] = pd.to_datetime(
    df_clean["registration_date"], format="%m/%Y", errors="coerce"
)

# Check how many rows failed to convert
invalid_dates = df_clean["registration_date"].isna().sum()
print(f"Invalid registration_date entries: {invalid_dates}")
```

```python
# Convert year to integer safely
df_clean.loc[:, "year"] = pd.to_numeric(
    df_clean["year"], errors="coerce"
).astype("Int64")

# Check for missing or strange values
print(f"Missing year values: {df_clean['year'].isna().sum()}")
print(f"Year range: {df_clean['year'].min()} to {df_clean['year'].max()}")
```

```python
from datetime import datetime

current_year = datetime.now().year
df_clean.loc[:, "car_age"] = current_year - df_clean["year"]
```
